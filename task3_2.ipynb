```
import numpy as np
import sys
sys.setrecursionlimit(100)
```

```
def g(x):
    return x+1
```

```
def bisection(xl,xu ,prev_xr):
  xr = xl+xu/2
  ea = abs(100 * (xr - prev_xr)/xr)
  x = np.linspace(xl,xu,11)
  if ea > es :
    if(g(xl)*g(xu)<0):
      prev_xr = xr
      return bisection(xl,xr,prev_xr)
    else:
      return bisection(xr,xl,prev_xr)
  else :
    print('END')
```

```
xl = 4
xu = 5
xr = xl+xu/2
prev_xr = 0
p = 4
es = 0.5*10**(2-p)

bisection(4,5,prev_xr)
```
